```python
# عندي list فيها اسماء منتجات فيها مسافات زيادة
# و capital & small letters مش مظبوطة
# المطلوب اني انضف الداتا باستخدام map و lambda

products = [" LAPTOP ", "phone ", " Tablet", "CAMERA "]
after = list(map(lambda x: x.strip().title(), products))
print(after)
```

```python
# هنا عندي list فيها درجات حرارة بالسيلزيوس
# وعايز احولها لفهرنهايت باستخدام map و lambda

celsius = [0, 10, 20, 30, 40]
fahrenheit = list(map(lambda c: (9/5)*c + 32, celsius))
print(fahrenheit)
```

```python
# هنا كل رقم في الليست عايزه اعمله عمليات
# الاول تربيع الرقم
# وبعدها اضيف 10 باستخدام lambda واحدة

nums = [1, 2, 3, 4, 5]
res = list(map(lambda n: n*n + 10, nums))
print(res)
```

```python
# عندي ليست كلمات وعايزة اطلع اول حرف واخر حرف من كل كلمة
# هستخدم map مع lambda
# واجيب الحروف عن طريق [0] و [-1]

words = ["python", "lambda", "programming", "map", "function"]
res = list(map(lambda w: (w[0], w[-1]), words))
print(res)
```

```python
marks = [[45, 80, 70], [90, 60, 100], [88, 76, 92]]

new = list(map(
    lambda row: list(map(lambda x: int(x * 1.05 + 0.5), row)),
    marks))
print(new)
```

```python
# normalization
# عندي ليست ارقام وعايزة اقسم كل رقم على اكبر رقم في الليست
# باستخدام map و lambda

numbers = [5, 10, 15, 20, 25]

min_val = min(numbers)
max_val = max(numbers)

normalized = list(
    map(lambda x: (x - min_val) / (max_val - min_val), numbers))

print(normalized)
```

```python
# جمل وكل جمله فيها كلمات
# عايزة اجيب طول كل كلمة
# باستخدام split و nested map

sentences = ["I love python", "map and lambda are useful"]
length = list(map(lambda s: list(map(lambda w: len(w), s.split())), sentences))
print(length)
```
